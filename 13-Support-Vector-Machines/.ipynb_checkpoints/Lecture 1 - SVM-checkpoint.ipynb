- Similar to how in communication systems we would distingish between two different types of signals (0, 1)
- We would project the signals to a different dimension, and then we can come up with regions that the signals would fall in

### Kernel
- We move the points to a higher dimension. So for example: 
<img src="kernel.png">

- Here we cannot come up with one hyperplane to distingish between the two categories, so we will project it the to x^2 dimension

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../DATA/mouse_viral_study.csv")
```

```python
df.info()
```

```python
sns.scatterplot(data=df, x="Med_1_mL", y="Med_2_mL", hue="Virus Present")

x = np.linspace(0, 10, 100)
y = -x + 11
plt.plot(x, y, "black")
```

```python
from sklearn.svm import SVC
```

```python
y = df["Virus Present"]
X = df.drop("Virus Present", axis=1)
```

```python
model = SVC(kernel="linear", C=1000)
```

```python
model.fit(X, y)
```

```python
from svm_margin_plot import plot_svm_boundary
```

```python
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel="linear", C=0.05)
model.fit(X, y)
```

```python
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel="rbf", C=1, gamma=0.000001)
model.fit(X, y)
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel="sigmoid")
model.fit(X, y)
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel="poly", degree=2)
model.fit(X, y)
plot_svm_boundary(model, X, y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC()
param_grid = {"C":[0.01,0.1, 1], 'kernel':["linear", "rbf"]}
```

```python
grid = GridSearchCV(svm, param_grid)
```

```python
grid.fit(X, y)
```

```python
grid.best_params_
```
