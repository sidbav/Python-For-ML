```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("country_table.csv")
```

```python
df = df.set_index("Countries")
```

```python
plt.figure(dpi=100)
sns.heatmap(data=df.drop("Life expectancy",axis=1),linewidth=0.5,annot=True,cmap="viridis",center=40)
```

```python
plt.figure(dpi=100)
sns.clustermap(data=df.drop("Life expectancy",axis=1),linewidth=0.5,annot=True,cmap="viridis",col_cluster=False)
```
