# Categorical Distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(data=df,x="level of education",hue="division")
df["division"].value_counts() # Probably a better way to do this
df["level of education"].value_counts()
```

```python
sns.barplot(data=df,x="level of education",y="salary",estimator=np.mean,hue="division")
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(10,8), dpi=300)
sns.violinplot(data=df, y="parental level of education", x="reading score", hue="test preparation course", split=True)
plt.legend(bbox_to_anchor=(1.25,0.5))
```

```python
plt.figure(figsize=(10,5),dpi=100)
sns.swarmplot(data=df, x="math score")
```
