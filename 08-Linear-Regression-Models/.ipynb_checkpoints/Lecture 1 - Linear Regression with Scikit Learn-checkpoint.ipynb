```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(16,6))

axes[0].plot(df['TV'],df['sales'],'o')
axes[0].set_ylabel("Sales")
axes[0].set_title("TV Spend")

axes[1].plot(df['radio'],df['sales'],'o')
axes[1].set_title("Radio Spend")
axes[1].set_ylabel("Sales")

axes[2].plot(df['newspaper'],df['sales'],'o')
axes[2].set_title("Newspaper Spend");
axes[2].set_ylabel("Sales")
plt.tight_layout();
```

```python
sns.pairplot(df)
```

```python
X = df.drop("sales",axis=1)
y = df["sales"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
len(df)
```

```python
len(X_train)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
help(LinearRegression)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
test_pre = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df["sales"].mean()
```

```python
sns.histplot(df["sales"])
```

```python
mean_absolute_error(y_test, test_pre) # Around 10% error
```

```python
np.sqrt(mean_squared_error(y_test, test_pre))
```

```python
# RSME allows to see if there are points where the error is off by a lot
```

## Residual Plots

- Allows us to be able to tell if the degree of our polynomial actually makes sense for the data given
- We will plot the equation that we calculate, and plot the difference between the two points and the 
- If we are the graph the distribution of the resduial values, we should close to a normal distribution
- The plot itself should not follow any pattern, should appear to be random

```python
test_res = y_test - test_pre
```

```python
test_res
```

```python
sns.scatterplot(x=y_test, y=test_res)
plt.axhline(y=0, color="red")
```

```python
sns.displot(test_res, bins=25, kde=True)
```

```python
import scipy as sp
```

```python
# Create a figure and axis to plot on
fig, ax = plt.subplots(figsize=(6,8),dpi=100)
# probplot returns the raw values if needed
# we just want to see the plot, so we assign these values to _
_ = sp.stats.probplot(test_res,plot=ax)
```

```python
final_model = LinearRegression()
```

```python
final_model.fit(X,y)
```

```python
final_model.coef_ # This is the beta coefficent for each column in the X data set
```

```python
X.head()
```

```python
y_hat = final_model.predict(X)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(16,6))

axes[0].plot(df['TV'],df['sales'],'o')
axes[0].plot(df['TV'],y_hat,'o',color='red')
axes[0].set_ylabel("Sales")
axes[0].set_title("TV Spend")

axes[1].plot(df['radio'],df['sales'],'o')
axes[1].plot(df['radio'],y_hat,'o',color='red')
axes[1].set_title("Radio Spend")
axes[1].set_ylabel("Sales")

axes[2].plot(df['newspaper'],df['sales'],'o')
axes[2].plot(df['radio'],y_hat,'o',color='red')
axes[2].set_title("Newspaper Spend");
axes[2].set_ylabel("Sales")
plt.tight_layout();
```
