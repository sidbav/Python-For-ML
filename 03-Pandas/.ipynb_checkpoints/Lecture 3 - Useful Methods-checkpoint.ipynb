```python
def last_four(number):
    return str(number)[-4:]
```

```python
last_four(1234455666)
```

```python
df["last_four"] = df["CC Number"].apply(last_four) 
# The apply method allows us to pass in a function, which we would like to run on all of the values in a column
```

```python
df
```

```python
df["total_bill"].mean()
```

```python
def yelp(price):
    if price < 10:
        return "$"
    elif price < 30:
        return "$$"
    else:
        return "$$$"
```

```python
df["yelp"] = df["total_bill"].apply(yelp)
```

```python
df
```

```python
def simple(num):
    return num*2
```

```python
simple(2)
```

```python
lambda num: num*2
```

```python
df["total_bill"].apply(simple)
```

```python
df["total_bill"].apply(lambda num:num*2)
```

```python
df.head()
```

```python
def quality(total_bill, tip):
    if (tip/total_bill) > .25:
        return "Generous"
    return "Other"
```

```python
quality(13, 34)
```

```python
lambda total_bill, tip:
```

```python
# This method beloow is used such that you are able to run apply on multiple columns at the same time
```

```python
df["quality"] = df[["total_bill", "tip"]].apply(lambda df: quality(df["total_bill"], df["tip"]), axis=1)
```

```python
df
```

```python
df["quality"] = np.vectorize(quality)(df["total_bill"], df["tip"])
```
