```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
image_as_array = mpimg.imread("../DATA/palm_tree.jpg")
```

```python
image_as_array.shape
```

```python
image_as_array
```

```python
plt.imshow(image_as_array)
```

```python
(h,w,c) = image_as_array.shape
```

```python
image_as_array_2d = image_as_array.reshape(h*w, c)
```

```python
image_as_array_2d
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=10)
```

```python
labels = model.fit_predict(image_as_array_2d)
```

```python
labels
```

```python
rgb = model.cluster_centers_.round(0).astype(int)
```

```python
rgb
```

```python
labels
```

```python
rgb[labels]
```

```python
quantized_image = np.reshape(rgb[labels], (h,w,c))
```

```python
plt.imshow(quantized_image)
```
