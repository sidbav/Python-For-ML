# Matplotlib SubPlots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 10, 11)
b = a**4
x = np.arange(0, 10)
y = x*2
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2,dpi=100)

axes[0][0].plot(a, b)
axes[1][0].plot(x, y)
axes[0][1].plot(x, y)
axes[1][1].plot(a, b)

axes[1][0].set_title("Hello")
axes[1][0].set_xlabel("X")
axes[1][0].set_ylabel("Y")

fig.suptitle("HELLO WORLD")
fig.tight_layout()
fig.savefig("my_subplots.png", bbox_inches="tight")
```
