# Matplotlib Figures

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1]) ## the first 2 values represent where the orgin of the axis will be 
# The last two values represent the width and height
x = np.arange(0, 10)
y = x*2
axes.plot(x, y)
```

```python
a = np.linspace(0, 10, 11)
```

```python
fig = plt.figure()
# Large Axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0, 5000)
axes1.set_title("Power of 4")
axes1.set_xlabel("a")
axes1.set_ylabel("b")

# Small Axes
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(a, b)
axes2.set_xlim(1,2)
axes2.set_ylim(0, 20)
axes2.set_title("Zoomed In")
axes2.set_xlabel("a")
axes2.set_ylabel("b")
```

```python
fig = plt.figure(dpi=100, figsize=(2,2))
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a,b)
fig.savefig("Newfigure.png", bbox_inches="tight")
```
