# Matplotlib Styling

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0, 1,1])
x = np.linspace(0, 11)
axes.plot(x, x,label="X vs X")
axes.plot(x, x**3,label="X vs X^3")
axes.legend()
```

```python
x = np.linspace(0, 11)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0, 1,1])
axes.plot(x,x,color="#000000",label="x vs x",lw=4,ls="-.")
axes.plot(x,x+2,color="purple", label="x vs x+2")
axes.plot(x,x+3, label="x vs x+3",marker="p",ms=3)
axes.legend()
```
