```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../DATA/rock_density_xray.csv")
```

```python
df.head()
```

```python
df.columns = ["Signal", "Density"]
```

```python
df
```

```python
sns.scatterplot(data=df, x="Signal", y="Density")
```

```python
X = df["Signal"].values.reshape(-1, 1)
y = df["Density"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train, y_train)
```

```python
preds = lr_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(y_test, preds)
```

```python
mean_squared_error(preds, y_test)
```

```python
np.sqrt(mean_squared_error(preds, y_test))
```

```python
preds
```

```python
plt.figure(figsize=(12, 8), dpi=200)
sns.scatterplot(data=df, x="Signal", y="Density")

signal_range = np.arange(0, 100)
preds = lr_model.predict(signal_range.reshape(-1, 1))
sns.scatterplot(x=signal_range, y=preds)
```

```python
def run_model(model,X_train,y_train,X_test,y_test):
    
    # Fit Model
    model.fit(X_train,y_train)
    
    # Get Metrics
    
    preds = model.predict(X_test)
    
    rmse = np.sqrt(mean_squared_error(y_test,preds))
    print(f'RMSE : {rmse}')
    
    # Plot results
    signal_range = np.arange(0,100)
    output = model.predict(signal_range.reshape(-1,1))
    
    
    plt.figure(figsize=(12,6),dpi=150)
    sns.scatterplot(x='Signal',y='Density',data=df,color='black')
    plt.plot(signal_range,output)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
pipe = make_pipeline(PolynomialFeatures(degree=7), LinearRegression())
```

```python
run_model(pipe, X_train, y_train, X_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
k_values = [1, 5, 10, 15, 20]

for n in k_values:
    model = KNeighborsRegressor(n_neighbors=n)
    run_model(model, X_train, y_train, X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor()
run_model(model, X_train, y_train, X_test, y_test)
```

```python
from sklearn.svm import SVR
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svr = SVR()
param_grid = {'C':[0.01,0.1,1,5,10,100,1000],'gamma':['auto','scale']}
```

```python
grid = GridSearchCV(svr, param_grid)
run_model(grid, X_train, y_train, X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(n_estimators=10)
```

```python
run_model(rfr, X_train, y_train, X_test, y_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor, AdaBoostRegressor
```

```python
model = AdaBoostRegressor()
```

```python
run_model(model, X_train, y_train, X_test, y_test)
```
